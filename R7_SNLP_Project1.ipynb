```python
# Considering fewer rows from the dataset for initial analysis

# Read the dataset

df=pd.read_csv('blogtext.csv', nrows=5000)
df
```

```python
# 2) Preprocessing on the Text column
import nltk 
import string 
import re 
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize

# Remove punctutations

df['text'] = df['text'].str.replace( '[^\w\s]' , '' )
```

```python
# Remove unwanted whitespaces
df['text'] = df['text'].str.strip()
```

```python
# Convert to Lowercase
df['text'] = df['text'].apply( lambda i : ' '.join( word.lower() for word in i.split()))
```

```python
# Remove stop words

stop = stopwords.words('english')
df['text'] = df['text'].apply( lambda t : " ".join( word for word in t.split() 
                                                       if word not in stop ) )
```

```python
# Remove numeric

df['text'] = df['text'].apply( lambda t : re.sub(r'\d+', '', t))
```

```python
df['text'][0]
```

```python
df
```

```python
# 3) Merging labels together

df2 = df.assign(labels = df.gender.astype(str) + ', ' +
  df.age.astype(str) + ', ' + df.topic.astype(str) + ', ' +
  df.sign.astype(str))

df2['labels'] = [x for x in df2['labels'].str.split(',') if x]
df2['labels']


```

```python
# consider only text and labels for model building

df2.drop(columns=['id','gender','age','topic','sign','date'])
```

```python
# 4) Separate features and labels for training and testing

X= df2['text']
y=df2['labels']

# split X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_train
```

```python
X_test
```

```python
# 5) Vectorize the features

# instantiate the vectorizer
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(ngram_range=(1, 2))

# Vectorize training and test set
X_train_dtm = vect.fit_transform(X_train)
X_test_dtm = vect.transform(X_test)
X_train_dtm
X_test_dtm

```

```python
# examine the vocabulary and document-term matrix together for train set
pd.DataFrame(X_train_dtm.toarray(), columns=vect.get_feature_names())
```

```python
# examine the vocabulary and document-term matrix together for test set
pd.DataFrame(X_test_dtm.toarray(), columns=vect.get_feature_names())
```

```python
# 6)  Count of each distnct label value and store in a dictionary

vec = CountVectorizer(tokenizer=lambda x: [i.strip() for i in x.split(',')], lowercase=False)
counts = vec.fit_transform(df2['labels'])
dict(zip(vec.get_feature_names(), counts.sum(axis=0).tolist()[0]))
```

```python
# 7) MultilabelBinarizer

from sklearn.preprocessing import MultiLabelBinarizer

multilabel_binarizer = MultiLabelBinarizer()
multilabel_binarizer.fit(y_train)

# transform train label
y_train = multilabel_binarizer.transform(y_train)
# transform test label
y_test = multilabel_binarizer.transform(y_test)
```

```python
multilabel_binarizer.inverse_transform(y_train)
```

```python
#8) OnevsRest on Logistic Regression for Model building

from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn import metrics

lr = LogisticRegression()
clf = OneVsRestClassifier(lr)

# fit model on train data
clf.fit(X_train_dtm, y_train)

# predict for the test set
y_pred = clf.predict(X_test_dtm)

```

```python
y_pred[0]
```

```python
# Accuracy for the model
print(metrics.accuracy_score(y_test, y_pred))
```

```python
# 9) Classification Report - F1 Score, Avg Precision and Avg Recall
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(metrics.accuracy_score(y_test, y_pred))
print(classification_report(y_test,y_pred))
```
