```python
# Colaboratve Filtering

from surprise import Dataset,Reader

reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(rating_1[['UserID', 'ProductID', 'Rating']], reader)
data
```

```python
# Split data to train and test in the Ratio of 7:3
from surprise.model_selection import train_test_split
trainset, testset = train_test_split(data, test_size=0.3,random_state=123)
type(trainset)
```

```python
# Training the Model

from surprise import KNNWithMeans
from surprise import accuracy
from surprise import Prediction

```

```python
# Using Pearson similarity
algo = KNNWithMeans(sim_options={'name': 'pearson', 'user_based': False})
algo.fit(trainset)
```

```python
# Evalute on test set
test_pred = algo.test(testset)

# compute RMSE
accuracy.rmse(test_pred)
```

```python
# using Cosine similarity
algo1= KNNWithMeans(sim_options={'name': 'cosine', 'user_based': False})
algo1.fit(trainset)
test_pred_1 = algo1.test(testset)
accuracy.rmse(test_pred_1)
```

```python
# Based on the similarity tests and RMSE observed, RMSE is better with Pearson than with Cosine similarity for the Item based collaborative filtering in this case
```

```python
# Summary

# Given dataset has the ratings given by easch user to the product purchased in the range of 1-5
# In the Popularity based recommendation, since it's common to all without any personalisation, each user will be recommended with the products that got the highest avergae ratings among the total users
# Products with the highest average rating is derived for both the train and test set
# In the Collaborative based filtering, the problem is to identify similarity among users who have given the similar ratings to the products. In the event, similar users will be recommended with the other products that they did not purchase.
# While building the model, the RMSE for the model is computed with both Cosine and Pearson distance.
# In general, lower the RMSE, better the model is, because the standard deviation of the data points in the pursuit of best fit will be minimum if RMSE is lower
# Since Pearson distance gives the lower RMSE, the model works better with this than with Cosine.
# Using the same similarity technique, the recommended products for each User is identified and noted.
```
