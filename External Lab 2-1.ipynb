### Create 500 random samples from normal distribution with mean 1 and  standard deviation 0.5

```python
import numpy as np;
mean, std = 1, 0.5;
sample = np.random.normal(mean, std, 500); ## 500 random normal distribution samples with mean 1 and standard deviation 0.5
print(sample);
```

###  Plot the histogram of the above normal distribution 

```python
plt.hist(sample);
plt.xlabel('Distribution');
plt.ylabel('No of Samples');
plt.grid(True)
plt.show();
```
