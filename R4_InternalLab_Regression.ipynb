### K - Means Clustering

### 4. Assign a dummy array called distortion

```python
# Intiate an array called Distortion

distortion=[]
```

You can later use this array to plot the elbow plot

```python
print(veh1.shape)
```

### 5. Calculating distortion - C distance

```python
from sklearn.cluster import KMeans
k_val=range(1,10);
a=[]
distortion=[]
k_cluster_Centers=[]


for i in k_val:
    km = KMeans(n_clusters=i, n_init=10, max_iter=300)
    #Fit on the data
    km.fit(veh1)
    #Add cluster information for each record in dataframe
    #veh1['veh1_' + str(i)] = km.predict(veh1)
    a.append(i)
    distortion.append(km.inertia_)
    #print(km.cluster_centers_)
    k_cluster_Centers.append(km.cluster_centers_)

distortion
```

Iterating values of k from 1 to 10 fit K means model
Using c distance - Get the measure for Sum of squares error.

```python
plt.plot(range(1,10), distortion, 'go-')
```

### 6. Plotting Elbow/ Scree Plot

```python
# Plot the Elbow graph using Distortion and Iteration

plt.plot(a, distortion)
plt.title("Scree Plot")
plt.xlabel("Iteration")
plt.ylabel("Distortion")
```

Use Matplotlib to plot the scree plot - Note: Scree plot plots distortion vs the no of clusters

### Find out the optimal value of K

```python
# From the elbow graph, we observe that the SSE value starts to form a pattern between 3 and 4
# Therefore, the optimal value of K would be 4
```

### Using optimal value of K - Cluster the data. 
Note: Since the data has more than 2 dimension we cannot visualize the data. As an alternative, we can observe the centroids and note how they are distributed across different dimensions

```python
# Build the model using Optimal K

km1 = KMeans(n_clusters=4)
km1.fit(veh1)
```

```python
a=km1.cluster_centers_

# Retrived from the loop above

a
```

## Hierarchical Clustering 

### 8. Variable creation

For Hierarchical clustering, we will create datasets using multivariate normal distribution to visually observe how the clusters are formed at the end

```python
a = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[100,])
b = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
c = np.random.multivariate_normal([10, 20], [[3, 1], [1, 4]], size=[100,])
```

```python
a
```

### 9. Combine all three arrays a,b,c into a dataframe

```python
combined=np.concatenate([a,b,c])
comb=pd.DataFrame(combined)
comb
```

### 10. Use scatter matrix to print all the 3 distributions

```python
# plot to show the distribution or density of values of two columns of the dataset combo

sns.scatterplot(comb[0], comb[1])
```

### 11. Find out the linkage matrix

```python
from scipy.cluster.hierarchy import cophenet, dendrogram, linkage
from scipy.spatial.distance import pdist

Z = linkage(comb, method='ward')
c, coph_dists = cophenet(Z , pdist(comb))
dist = pd.DataFrame(list(coph_dists), columns=['clusterDist'])
dist['ecludianDist'] = pdist(comb)
dist
```

Use ward as linkage metric and distance as Eucledian

### 12. Plot the dendrogram for the consolidated dataframe

```python
fig = plt.figure(figsize=(25, 10))
z=dendrogram(Z)
plt.show()
```

### 13. Recreate the dendrogram for last 12 merged clusters 

```python
fig = plt.figure(figsize=(25, 10))
a=dendrogram(Z, p=12,truncate_mode='lastp')
```

Hint: Use truncate_mode='lastp' attribute in dendrogram function to arrive at dendrogram 

### 14. From the truncated dendrogram, find out the optimal distance between clusters which u want to use an input for clustering data

```python
# On observing the above dendrogram and also the range of values for no of clusters 3, 
# we come to know that the optimal distance is close to 50 (approx node distance - min -25, max=75)
```

### 15. Using this distance measure and fcluster function to cluster the data into 3 different groups

```python
# Using distance and Fcluster to form a flat clcuster
import scipy
clusters = scipy.cluster.hierarchy.fcluster(Z, t=3, criterion='distance')
clusters
```

### Use matplotlib to visually observe the clusters in 2D space 

```python
# Plot to visualise the column values versus the observed cluster for the given data

plt.scatter(comb[0], comb[1], c=clusters)
```
