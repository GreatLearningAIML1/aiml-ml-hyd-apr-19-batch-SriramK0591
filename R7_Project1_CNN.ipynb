```python
# 1) Merging the given train data of each class together
x_train = []
y_train = []
import cv2
for i in os.listdir():
    print(i)
    if (os.path.isdir(i)):
            for j in os.listdir(i):
                try:
                    dummy = cv2.imread('/content/train/' + i + "/" + j)
                    dummy = cv2.resize(dummy,(128,128))
                    x_train.append(dummy)
                    y_train.append(i)
                except Exception as e:
                    print(e)
```

```python
# Encoding the class variable

import pandas as pd
y_train = pd.get_dummies(y_train)
```

```python
# Converting X and Y sets to Arrays

import numpy as np
x_train = np.array(x_train)
y_train = np.array(y_train)
```

```python
# Splitting the given data into Train and Validation set in the ratio of 4:1

from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=2)
```

```python
print (len(x_train))
print (len(x_val))
```

```python
# Reshaping the X set

x_train = x_train.reshape(x_train.shape[0],128,128,3)
x_val = x_val.reshape(x_val.shape[0],128,128,3)
```

```python
# Printing the shapes of train and validation sets

print (x_train.shape)
print (x_val.shape)
print (y_train.shape)
print (y_val.shape)
```

```python
import numpy as np
import keras
from keras.datasets import cifar10, mnist
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Flatten, Reshape
from keras.layers import Convolution2D, MaxPooling2D
from keras.utils import np_utils
import pickle
from matplotlib import pyplot as plt
import seaborn as sns
from keras.layers import BatchNormalization
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.optimizers import adam
```

```python
# Initialize and Build the Model


TRAIN = False
BATCH_SIZE = 32
EPOCHS = 10

# Define the Type of Model
model1 = Sequential()
# 1st Conv Layer
model1.add(BatchNormalization(input_shape = (128,128,3)))
model1.add(Convolution2D(32, (3,3), activation ='relu', input_shape = (128, 128, 3)))
model1.add(Activation('relu'))

# 2nd Conv Layer
model1.add(Convolution2D(32, 3, 3))
model1.add(Activation('relu'))

# Fully Connected Layer
model1.add(Flatten())
model1.add(Dense(128))
model1.add(Activation('relu'))

# Dense Layer
model1.add(Dense(12))
model1.add(Activation('softmax'))

# Loss and Optimizer
model1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
# Train the model
model1.fit(x_train, y_train, batch_size=BATCH_SIZE, nb_epoch=EPOCHS, 
              validation_data=(x_val, y_val))
    


```

```python
# Optimize the model
BATCH_SIZE = 32
EPOCHS = 10

# Define Model
model2 = Sequential()

# 1st Conv Layer
model2.add(Convolution2D(32, 3, 3, input_shape=(128, 128, 3)))
model2.add(Activation('relu'))

# 2nd Conv Layer
model2.add(Convolution2D(32, 3, 3))
model2.add(Activation('relu'))

# Max Pooling
model2.add(MaxPooling2D(pool_size=(2,2)))
    
# Dropout
model2.add(Dropout(0.25))

# Fully Connected Layer
model2.add(Flatten())
model2.add(Dense(128))
model2.add(Activation('relu'))
model2.add(Dropout(0.5))

# Dense Layer
model2.add(Dense(12))
model2.add(Activation('softmax'))

# Loss and Optimizer
model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model2.fit(x_train, y_train, batch_size=BATCH_SIZE, nb_epoch=EPOCHS, 
              validation_data=(x_val, y_val))
    
```

```python
# Improving the model further

model4 = Sequential()
model4.add(BatchNormalization(input_shape = (128,128,3)))

# First Layer
model4.add(Convolution2D(32, (3,3), activation ='relu', input_shape = (128, 128, 3))) 
model4.add(MaxPooling2D(pool_size=2))
model4.add(Dropout(0.8))

#Second layer
model4.add(Convolution2D(filters=64, kernel_size=4, padding='same', activation='relu'))
model4.add(MaxPooling2D(pool_size=2))
model4.add(Dropout(0.8))

#Third layer
model4.add(Convolution2D(filters=128, kernel_size=3, padding='same', activation='relu'))
model4.add(MaxPooling2D(pool_size=2))
model4.add(Dropout(0.8))

#Fourth layer
model4.add(Convolution2D(filters=128, kernel_size=2, padding='same', activation='relu'))
model4.add(MaxPooling2D(pool_size=2))
model4.add(Dropout(0.8))
model4.add(Flatten()) 

# fully connected layer
model4.add(Dense(units=128,activation = 'relu'))

model4.add(Dense(units = 64, activation = 'relu'))
model4.add(Dropout(0.3))

model4.add(Dense(units = 32, activation = 'relu'))
model4.add(Dropout(0.3))

model4.add(Dense(units = 12, activation = 'softmax')) 

optimizer = adam(lr=0.001)
model4.compile(optimizer='adam', loss = 'categorical_crossentropy',metrics = ['accuracy'])

batch_size = 64
nb_epochs = 10
history = model4.fit(x_train,y_train,
                    epochs=20, 
                    validation_data=(x_val,y_val),
                    verbose = 1,
                    initial_epoch=0)
```

```python
# Improving the model further

model3 = Sequential()
model3.add(BatchNormalization(input_shape = (128,128,3)))
# First layer
model3.add(Convolution2D(32, (3,3), activation ='relu', input_shape = (128, 128, 3))) 
model3.add(MaxPooling2D(pool_size=2))

#Second layer
model3.add(Convolution2D(filters=64, kernel_size=4, padding='same', activation='relu'))
model3.add(MaxPooling2D(pool_size=2))

# Third layer
model3.add(Convolution2D(filters=128, kernel_size=3, padding='same', activation='relu'))
model3.add(MaxPooling2D(pool_size=2))

#Fourth layer
model3.add(Convolution2D(filters=128, kernel_size=2, padding='same', activation='relu'))
model3.add(MaxPooling2D(pool_size=2))
model3.add(Flatten()) 

# fully connected layer with different units
model3.add(Dense(units=128,activation = 'relu'))
model3.add(Dense(units = 64, activation = 'relu'))
model3.add(Dense(units = 32, activation = 'relu'))
model3.add(Dense(units = 12, activation = 'softmax')) 

optimizer = adam(lr=0.001)
model3.compile(optimizer='adam', loss = 'categorical_crossentropy',metrics = ['accuracy'])

batch_size = 64
nb_epochs = 10
history = model3.fit(x_train,y_train,
                    epochs=20, 
                    validation_data=(x_val,y_val),
                    verbose = 1,
                    initial_epoch=0)
```

```python
# Accuracy for Train data using each Model built

# Model1
print ("Accuracy of Model1 is",model1.evaluate(x_train,y_train));

# Model2
print ("Accuracy of Model2 is",model2.evaluate(x_train,y_train));

# Model3
print ("Accuracy of Model3 is",model3.evaluate(x_train,y_train));

# Model4
print ("Accuracy of Model4 is",model4.evaluate(x_train,y_train));

# Inference:
# Model3 gives a better accuracy of over 99% of all the other models
```

```python
# Accuracy of Validation data using each Model built

# Model1
print ("Accuracy of Model1 is",model1.evaluate(x_val,y_val));

# Model2
print ("Accuracy of Model2 is",model2.evaluate(x_val,y_val));

# Model3
print ("Accuracy of Model3 is",model3.evaluate(x_val,y_val));

# Model4
print ("Accuracy of Model4 is",model4.evaluate(x_val,y_val));

# Inference:
# Model3 gives a better accuracy of over 81% of all the other models
```
