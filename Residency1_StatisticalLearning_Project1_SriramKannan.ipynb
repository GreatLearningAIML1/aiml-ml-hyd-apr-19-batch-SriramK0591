```python
## Importing the libraries for performing analysis over the dataset

import numpy as np;
import pandas as pd;
```

```python
## read the data from the csv dataset
## First row is a blank, hence setting the second row as a header

laliga = pd.read_csv(r'C:\Users\sriram.kannan\Desktop\PGP-AIML\Residency1_Assignments\Project\Laliga.csv', sep=',', header=1);
print (laliga);



```

```python
## 1) Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

laliga = laliga.replace('-',0);
print (laliga);
```

```python
## 2) Print all the teams which have started playing between 1930-1980.

laliga_t = laliga.loc[laliga['Debut'].str[:4].between ('1930', '1980', inclusive=True)];

print (laliga_t); 



```

```python
## 3) Print the list of teams which came Top 5 in terms of points

## To sort, column 'Points' is changed from Object datatype to Int

laliga['Points']=laliga['Points'].astype('int64');

desc = laliga.sort_values(by='Points', ascending=False);

print (desc.iloc[0:5].Team); 

```

```python
## 4) Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.

## Goal_diff_count = GoalsFor - GoalsAgainst

## Converting datatypes of GoalsFor and GoalsAgainst from Object to Int for computation

laliga['GoalsFor']=laliga['GoalsFor'].astype('int64');
laliga['GoalsAgainst']=laliga['GoalsAgainst'].astype('int64');

def Goal_diff_count (a,b,c):
    team_goaldiff= dict(zip(a,b-c));
    print (team_goaldiff); ## prints the team vs goal difference in a dict structure
    print ("Team with maximum goal difference is", max(team_goaldiff, key=team_goaldiff.get)); #  prints the team that has the highest goal difference
    print ("Team with minimum goal difference is", min(team_goaldiff, key=team_goaldiff.get)); # prints the team that has the lowest goal difference
    
    
Goal_diff_count(laliga['Team'], laliga['GoalsFor'], laliga['GoalsAgainst']); ## calling a function using input parameters from the dataframe





    
    
    
    
    
```

```python
## 5)  Create a new column with name “Winning Percent” and append it to the data set (5 points)
## Percentage of Winning = (GamesWon / GamesPlayed)*100
## If there are any numerical error, replace it with 0%
## Print the top 5 teams which has the highest Winning percentage

# converting the columns 'GamesWon' and 'GamespLayed' into Int from Object datatype for computation

laliga['GamesWon']=laliga['GamesWon'].astype('int64');
laliga['GamesPlayed']=laliga['GamesPlayed'].astype('int64');

laliga['Winning Percent'] = laliga['GamesWon']*100/ laliga['GamesPlayed'];

## Fills Null or Numerical errors with 0
laliga['Winning Percent'].fillna(0, inplace=True);

## Top 5 teams with the highest winning percentage

laliga1 = laliga.sort_values(by='Winning Percent', ascending=False);
laliga1 = laliga1.iloc[0:5,].Team;

print (laliga);
print (laliga1);
```

```python
## 6) Group teams based on their “Best position” and print the sum of their points for all positions

total_points_per_pos = laliga.groupby(['BestPosition']).sum().Points;

print (total_points_per_pos);
```
